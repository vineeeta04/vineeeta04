```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
product_name=[]
reviews=[]
links=[]
for i in range(2,12):
    url="https://www.flipkart.com/search?p%5B%5D=facets.brand%255B%255D%3Drealme&sid=tyy%2F4io&sort=recency_desc&ctx=eyJjYXJkQ29udGV4dCI6eyJhdHRyaWJ1dGVzIjp7InRpdGxlIjp7Im11bHRpVmFsdWVkQXR0cmlidXRlIjp7ImtleSI6InRpdGxlIiwiaW5mZXJlbmNlVHlwZSI6IlRJVExFIiwidmFsdWVzIjpbIkxhdGVzdCBTYW1zdW5nIG1vYmlsZXMgIl0sInZhbHVlVHlwZSI6Ik1VTFRJX1ZBTFVFRCJ9fX19fQ%3D%3D&wid=1.productCard.PMU_V2_1&page="+str(i)
    r=requests.get(url)

    soup=BeautifulSoup(r.text,"lxml")
    box=soup.find("div",class_="_1YokD2 _3Mn1Gg")
    names=box.find_all("div",class_="_4rR01T")
    for i in names:
        name=i.text
        product_name.append(name)
    print("product name")
    print(len(product_name))

    reviews_data=box.find_all("div",class_="_3LWZlK")
    for i in reviews_data:
        name=i.text
        reviews.append(name)
    print("reviews")
    print(len(reviews))
    content=BeautifulSoup(r.content,'html.parser')
    data=content.find_all("div",{'class':'_2kHMtA'})
    start_link="https://www.flipkart.com"
    for items in data:
        rest_link=items.find('a')['href']
        links.append(start_link+rest_link)
        
    print("links")
    print(len(links))
```

```python
df=pd.DataFrame({"Product name":product_name,"Links of products":links,"Reviews":reviews})
df.to_csv("C:/Users/prafu/OneDrive/Desktop/python aba datasets/realme_mobiles.csv")
```
